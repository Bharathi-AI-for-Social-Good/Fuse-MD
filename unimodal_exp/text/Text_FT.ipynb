```python
import pandas as pd
from simpletransformers.classification import ClassificationModel
from sklearn.model_selection import train_test_split
```

```python
model = "google/muril-base-cased"
model_name = model.split("/")[1]


lang = "tamil"
lr =  5e-5
batch = 16
epochs = 5
model_save = str(model_name) + "_" + str(epochs) + "_" + str(lr)
output_path = "/content/drive/MyDrive/work/Meme papers/misogyny meme work/misosyny meme work/saved_models/"+ lang +"/"+model_save
model_save
```

```python
# Define the model
model = ClassificationModel(
    "bert",
    model,  # You can use a different BERT model if needed
    num_labels=2,  # 2 labels: 0 for Not a Hate Speech, 1 for Hate Speech
    use_cuda=True,
    args={"output_dir":output_path,  \
          "best_model_dir":output_path+"/best_model/",  \
          "overwrite_output_dir": True, "reprocess_input_data": True, "num_train_epochs": epochs, "train_batch_size":batch,"save_model_every_epoch": True ,  \
          "logging_steps":50, "learning_rate": lr},  \

)

# Train the model
model.train_model(train_df)

```
