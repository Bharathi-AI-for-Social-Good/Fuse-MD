```python
train_csv = pd.read_csv('/content/drive/MyDrive/work/Meme papers/misogyny meme work/Misogyny_meme_work1/training_data/tamil/train/train.csv')
test_csv = pd.read_csv('/content/drive/MyDrive/work/Meme papers/misogyny meme work/Misogyny_meme_work1/training_data/tamil/test/test.csv')
```

```python
trainX = []
trainY = []

for i,j in zip(train_csv['image_id'],train_csv['labels']):
  trainY.append(str(j))
  image = load_img("/content/drive/MyDrive/work/Meme papers/misogyny meme work/Misogyny_meme_work1/training_data/tamil/train/"+str(i)+".jpg", target_size=(224, 224))
  image = tf.convert_to_tensor(image, dtype="float32")
  image = preprocess_input(image)
  trainX.append(image)

```

```python
trainX = tf.convert_to_tensor(trainX, dtype="float32")
trainY = tf.keras.utils.to_categorical(trainY, num_classes=2)
trainY = tf.convert_to_tensor(trainY)

# perform one-hot encoding on the labels
# lb = LabelBinarizer()
# trainY = lb.fit_transform(trainY)
# trainY = to_categorical(trainY)
```

```python

testX = []
testY = []

for i,j in zip(test_csv['image_id'],test_csv['labels']):
  testY.append(str(j))
  image = load_img("/content/drive/MyDrive/work/Meme papers/misogyny meme work/Misogyny_meme_work1/training_data/tamil/test/"+str(i)+".jpg", target_size=(224, 224))
  image = tf.convert_to_tensor(image, dtype="float32")
  image = preprocess_input(image)
  testX.append(image)
```

```python
testX = tf.convert_to_tensor(testX, dtype="float32")
testY = tf.convert_to_tensor(tf.keras.utils.to_categorical(testY, num_classes=2))
```

```python
vgg19_model = tf.keras.applications.vgg19.VGG19(weights='imagenet', include_top=False, input_tensor=Input(shape=(224, 224, 3)))
```

```python
vgg19_model.summary()
```

```python
# construct the head of the model that will be placed on top of the
# the base model
headModel = vgg19_model.output
headModel = AveragePooling2D((7, 7),padding="same")(headModel)
headModel = Flatten(name="flatten")(headModel)
headModel = Dense(128, activation="relu")(headModel)
headModel = Dropout(0.5)(headModel)
headModel = Dense(2, activation="softmax")(headModel)
```

```python
model = Model(inputs=vgg19_model.input, outputs=headModel)
```

```python
# for layer in vgg19_model.layers:
#     layer.trainable = False
```

```python
model.summary()
```

```python
learning_rate=5e-5
Epochs=10
BS=16

# compile our model
print("[INFO] compiling model...")
opt = Adam(learning_rate=learning_rate)
model.compile(loss="binary_crossentropy", optimizer=opt,
metrics=["accuracy"])

```

```python
trainY
```

```python
H = model.fit(trainX, trainY,
    batch_size=BS,
	steps_per_epoch=len(trainX) // BS,
	validation_data=(testX, testY),
	validation_steps=len(testX) // BS,
	epochs=Epochs)
```

```python

# make predictions on the testing set
print("[INFO] evaluating network...")
predIdxs = model.predict(testX, batch_size=BS)

# for each image in the testing set we need to find the index of the
# label with corresponding largest predicted probability
predIdxs = np.argmax(predIdxs, axis=1)

# show a nicely formatted classification report
print(confusion_matrix(np.argmax(testY, axis=1), predIdxs))
print(classification_report(np.argmax(testY, axis=1), predIdxs, digits=5, target_names=['misogyny','notmisogyny']))
```
